# Projet
___

- Créer une application déployable avec `snowflake CLI`.

    - S'assurer que les appels à `snowflake CLI` soient livrés avec le projet.

- S'assurer que le modèle `RBAC` est bien mis en place.

- Créer tout le code de transformation et de logging dans le schéma `COMMON` (tables de logging / procédures / fonctions / tasks).

- Ajouter une task `ArriveesTardives` pour identifier les données qui sont arrivées trop tard (i.e. ont plus de 5 jours de retard) et les exclure du processing.

- Enregistrer tout ceci dans un `repo GitHub`, ajouter un `README` pour expliquer ce qu vous avez créé et nous envoyer le lien pour revue :).
